```python
import pandas as pd

df = pd.read_csv('books.csv')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(df.isna().transpose(), cbar=False, ax=ax)
plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.show()
```

```python
import numpy as np

df['Missing_Description'] = np.where(df['description'].isna(), 1, 0)
df['Age_of_Book'] = 2025 - df['published_year']
```

```python
columns_of_interest = ['Missing_Description', 'Age_of_Book', 'num_pages', 'average_rating']
correlation_matrix = df[columns_of_interest].corr(method='spearman')
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")
heatmap.set_title('Correlation Heatmap')
plt.show()
```

```python
missing_book = df[ ~df['description'].isna() &
    ~df['published_year'].isna() &
    ~df['average_rating'].isna() &
    ~df['num_pages'].isna() &
    ~df['ratings_count'].isna() ]
missing_book
```

```python
missing_book['categories'].value_counts().reset_index().sort_values(by="count", ascending=False)
```

```python
missing_book.loc[:, 'length_of_description'] = missing_book['description'].str.split().str.len()
```

```python
missing_book_25 = missing_book[missing_book['length_of_description'] >= 25]
missing_book_25
```

```python
missing_book_25.loc[:, 'title/subtitle'] = (
    np.where(missing_book_25['subtitle'].isna, missing_book_25['title'], missing_book_25[['title','subtitle']].astype('str').agg(': '.join, axis=1))
)
```

```python
missing_book_25
```

```python
missing_book_25.loc[:, 'tagged_description'] = missing_book_25[['isbn13', 'description']].astype('str').agg(' '.join, axis=1)
```

```python
(
    missing_book_25
    .drop(['subtitle', 'length_of_description', 'Missing_Description', 'Age_of_Book'], axis=1)
    .to_csv('books_cleaned.csv', index=False)

)
```
