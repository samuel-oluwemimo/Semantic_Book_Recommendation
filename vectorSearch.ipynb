```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain.vectorstores import Chroma
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv('books_cleaned.csv')
books.head()
```

```python
books['tagged_description']
```

```python
books["tagged_description"].to_csv("tagged_description.txt",
                                   lineterminator="\n",
                                   index=False,
                                   header=False)
```

```python
raw_documents = TextLoader("tagged_description.txt",  encoding="utf-8").load()
```

```python
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

huggingface_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

db_books = Chroma.from_documents(
    documents,
    embedding=huggingface_embeddings
)
```

```python
query = 'a book about teenage love'
docs = db_books.similarity_search(query, k=5)
docs
```

```python
books[books['isbn13'] == int(docs[0].page_content.strip('"').split()[0])]
```

```python
def retrieve_semantic_recommendations(query, k=5) -> pd.DataFrame:
    """
    Retrieve semantic recommendations based on the query.

    Args:
        query (str): The query string.
        k (int): The number of recommendations to retrieve.

    Returns:
        pd.DataFrame: DataFrame containing the recommended books.
    """
    docz = db_books.similarity_search(query, k=k)
    recommended_books = []

    for doc in docz:
        isbn = int(doc.page_content.strip('"').split()[0])
        book = books[books['isbn13'] == isbn]
        recommended_books.append(book)

    return pd.concat(recommended_books, ignore_index=True)
```

```python
retrieve_semantic_recommendations(query='a book about humility', k=5)
```
